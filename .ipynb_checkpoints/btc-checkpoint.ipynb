### Write Data Cleaning Functions

```python
def snake_to_norm(snake):
    lst = snake.split("_")
    for i in range(len(lst)):
        lst[i] = lst[i].title()
    return " ".join(lst)

def norm_to_snake(norm):
    lst = norm.split(" ")
    for i in range(len(lst)):
        lst[i] = lst[i].lower()
    return "_".join(lst)
```

### Transaction Fees over Time (Before Clean)

```python
# Set the graph style
plt.style.use('default')

# Adjust the figure size
plt.figure(dpi=200)

# Input the column and index arguments
plt.plot(btc_df['height'],
         btc_df['avg_trans_fee_per_trans'],
         color='#C4E2FFFF'
        )    

# Change the names of the x and y axis
plt.xlabel('Block Height')
plt.ylabel('Transaction Fee (BTC)')

# Adding in the title
plt.title("Transaction Fees (Before Clean)")

# Save the image
plt.savefig('transaction_fee1.png')

# Display the Graph
plt.show()
```

### Clean Transaction Fee Data

```python
# Remove all the outlier the column 'avg_trans_fee_per_trans'
adj_vals = btc_df['avg_trans_fee_per_trans'][:]
mean = adj_vals.mean()
std = adj_vals.std()

# Remove values that have a z-score > 3
for i in range(len(adj_vals)):
    if abs((adj_vals[i]-mean)/std)>3:
        adj_vals.drop([i],inplace=True)
        
# Compare the before and after
print(btc_df['avg_trans_fee_per_trans'].describe())
print("Length before outlier removal:",len(btc_df['avg_trans_fee_per_trans']))
print()
print(adj_vals.describe())
print("Length after outlier removal:",len(adj_vals))
```

### Graph Transaction Fees over Time (After Clean)

```python
# Set the graph style
plt.style.use('default')

# Adjust the figure size
plt.figure(dpi=200)

# Input the column and index arguments
plt.plot(adj_vals.index,
         adj_vals,
         color='#C4E2FFFF'
        )    

# Change the names of the x and y axis
plt.xlabel('Block Height')
plt.ylabel('Transaction Fee (BTC)')

# Adding in the title
plt.title("Transaction Fees (After Clean)")

# Save the image
plt.savefig('transaction_fee2.png')

# Display the Graph
plt.show()
```

#### Analyse Price Data:
1) We need to first import price data using the YahooFinancials package

```python
cryptocurrencies = ['BTC-USD', 'ETH-USD', 'BNB-USD']
yf = YahooFinancials(cryptocurrencies)
crypto = yf.get_historical_price_data("2009-01-03", "2022-01-08", "daily")
```

2) Write the data to a new .csv file and read the new data we've collected

```python
def writeData():
    # Open file object in binary write mode
    file_object = open('crypto_prices','wb')

    # Source the destination
    pk.dump(crypto,file_object)
    file_object.close()

def readData():
    file_object = open('crypto_prices','rb')
    crypto = pk.load(file_object)
    file_object.close()
    return crypto

def main():
    #writeData()
    print(readData())

main()
```
