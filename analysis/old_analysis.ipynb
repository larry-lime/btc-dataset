### Write Data Cleaning Functions

```python
def snake_to_norm(snake):
    lst = snake.split("_")
    for i in range(len(lst)):
        lst[i] = lst[i].title()
    return " ".join(lst)

def norm_to_snake(norm):
    lst = norm.split(" ")
    for i in range(len(lst)):
        lst[i] = lst[i].lower()
    return "_".join(lst)
```
